```python
def ab(x):
    return  2*x + 1
```

```python
def gd(ig,df,multiplier=0.01,precessor=0.001,max_iter=100):
    
    new_x=ig
    
    x_list=[new_x]
    
    slope_list=[df(new_x)]
    
    for n in range(max_iter):
        
        old_x=new_x
        
        gradient=df(new_x)
        
        new_x=old_x-multiplier*gradient
        
        step_size=abs(new_x - old_x)
        
        x_list.append(new_x)
        
        slope_list.append(df(new_x))
        
        if step_size < precessor:
            
            break
        return new_x,x_list,slope_list
        
    
    
```

```python
x_2 = np.linspace(-3,3,1000)
```

```python
def g(x):
    
    return x**4 - 4*x**2 +5
```

```python

def dg(x):
    
    return 4*x**3 - 8*x
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier = 0.01, precision=0.001,  max_iter=100):
    
    
    
    new_x = initial_guess
    
    x_list = [new_x]
    
    slope_list = [derivative_func(new_x)]
    
       
    for n in range(max_iter):
        
        
        previous_x = new_x
        
        gradient = derivative_func(previous_x)
        
        new_x = previous_x - multiplier * gradient
        
        
        step_size = abs(new_x - previous_x)
        
        
        x_list.append(new_x)
    
        
        slope_list.append(derivative_func(new_x))
        
        
        if step_size < precision:
                
                    break 
    
    
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg, initial_guess=1.9, precision=0.0000001)
```
