```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 -4*x**2 + 5
```

```python
x_2=np.linspace(-3,3,1000)
```

```python
def df(x):
    return 4*x**3+ 8*x
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
plt.title("parabola",fontsize=25)
plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)
#values=np.array(x_list)
#plt.scatter(x_list,f(values),color="green",s=100,alpha=0.5)
plt.grid()
plt.plot(x_2,f(x_2))

plt.subplot(1,2,2)
plt.title("stright line",fontsize=25)
plt.xlabel('x',fontsize=18)
plt.ylabel('df(x)',fontsize=18)
#plt.scatter(x_list,slope,color="green",s=100,alpha=0.5)
plt.grid()
plt.plot(x_2,df(x_2))
plt.show()
```

```python
new_x = 2

previous_x = 2

gamma = 0.01

x_list = [new_x]

slope_list = [df(new_x)]

for n in range(1000):
    
    previous_x = new_x
    new_x = previous_x - gamma*df(previous_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
plt.title("parabola",fontsize=25)
plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)
plt.xlim(-2,2)
plt.ylim(-0.5,5.5)
values=np.array(x_list)
plt.scatter(x_list,f(values),color="green",s=100,alpha=0.5)
plt.grid()
plt.plot(x_2,f(x_2))

plt.subplot(1,2,2)
plt.title("stright line",fontsize=25)
plt.xlabel('x',fontsize=18)
plt.ylabel('df(x)',fontsize=18)
plt.xlim(-2,2)
plt.ylim(-10,10)
plt.scatter(x_list,slope_list,color="red",s=100,alpha=0.5)
plt.grid()
plt.plot(x_2,df(x_2))
plt.show()
```
