```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + x + 1
```

```python
x = np.linspace(start = -3, stop = 3, num = 200)
```

```python

def df(x):
    return 2*x + 1
```

```python
new_x=2.1
old_x=2
gamma=0.1

x_list=[new_x]
slope_list=[df(new_x)]

for n in range(75):
    old_x= new_x
    new_x=old_x-gamma*df(old_x)
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
print(round(new_x,2))    
    
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
plt.title("parabola",fontsize=25)
plt.xlabel('x',fontsize=18)
plt.ylabel('f(x)',fontsize=18)
values=np.array(x_list)
plt.scatter(x_list,f(values),color="green",s=100,alpha=0.5)
plt.grid()
plt.plot(x,f(x))

plt.subplot(1,2,2)
plt.title("parabola",fontsize=25)
plt.xlabel('x',fontsize=18)
plt.ylabel('df(x)',fontsize=18)
plt.scatter(x_list,slope_list,color="green",s=100,alpha=0.5)
plt.grid()
plt.plot(x,df(x))
plt.show()
```
