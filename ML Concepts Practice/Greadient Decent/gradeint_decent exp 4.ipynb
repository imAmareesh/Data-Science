```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
x_1=np.linspace(-2.5,2.5,1000)
```

```python
def h(x):
    return x**5-2*x**4+2

def dh(x):
    return 5*x**4-8*x**3
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1,2,1)
plt.title("cost function",fontsize=17)
plt.plot(x_1,h(x_1))
plt.subplot(1,2,2)
plt.title("derivative cost",fontsize=17)
plt.plot(x_1,dh(x_1))
plt.show()
```

```python
def gd(df, ig, multiplier = 0.01, precision=0.001,max_iter=72):
    
    
    
    new_x = ig
    
    x_list = [new_x]
    
    slope_list = [df(new_x)]
    
       
    for n in range(max_iter):
        
        
        previous_x = new_x
        
        gradient = df(previous_x)
        
        new_x = previous_x - multiplier * gradient
        
        
        step_size = abs(new_x - previous_x)
        
        
        x_list.append(new_x)
    
        
        slope_list.append(df(new_x))
        
        
        if step_size < precision:
                
                    break 
    
    
    return new_x, x_list, slope_list
```

```python
local_min, list_x, deriv_list = gd(df=dh,multiplier = 0.02, ig=-0.2, precision=0.01)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)
plt.grid()
plt.xlim(-1.2, 2.5)
plt.ylim(-2, 4)
plt.plot(x_1, h(x_1), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, h(np.array(list_x)), color='red', s=100, alpha=0.6)

plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)
plt.plot(x_1, dh(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)

plt.show()
```

```python
print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
print('Number of steps: ', len(list_x))
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier = 0.01, precision=0.001,  max_iter=100):
    
    
    
    new_x = initial_guess
    
    x_list = [new_x]
    
    slope_list = [derivative_func(new_x)]
    
       
    for n in range(max_iter):
        
        
        previous_x = new_x
        
        gradient = derivative_func(previous_x)
        
        new_x = previous_x - multiplier * gradient
        
        
        step_size = abs(new_x - previous_x)
        
        
        x_list.append(new_x)
    
        
        slope_list.append(derivative_func(new_x))
        
        
        if step_size < precision:
                
                    break 
    
    
    return new_x, x_list, slope_list
```

```python
x_2 = np.linspace(-2, 2, 1000)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)
plt.grid()
plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.8)
plt.scatter(list_x, g(np.array(list_x)), color='red', s=100, alpha=0.6)




plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.8)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.6)




plt.show()
```

```python
def gradient_descent( initial_guess,  multiplier, precision, max_iter):
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [dg(new_x)]

    for n in range(max_iter):
        previous_x = new_x 
        new_x = previous_x - multiplier * dg(previous_x)
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(dg(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
n = 100
local_min, list_x, deriv_list = gradient_descent(initial_guess=1.9, multiplier = 0.25, precision =0.0001, max_iter = n) 
# Plot function and derivative and scatter plot side by side

plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.grid()
plt.xlim(-2,2)
plt.ylim(-0.5, 6.5)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)

plt.plot(x_2, g(x_2), color='blue', linewidth=3, alpha=0.5)
plt.scatter(list_x, g(np.array(list_x)), color='green', s=100, alpha=0.9)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.plot(x_2, dg(x_2), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.5)

plt.show()
   
print('Number of steps:', len(list_x))
print('Local min occurs at:', local_min)
```
